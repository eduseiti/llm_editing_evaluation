```python
fig = plt.figure(figsize=[15, 10])

fig.suptitle("Question scores", y=0.91, fontsize='xx-large')

plt.grid()

for question in ordered['question'].unique():
    q = ordered[ordered['question'] == question][['round', 'mean_score', 'max', 'min']]

    line, = plt.plot(q['round'], q['mean_score'], alpha=0.8)
    # plt.fill_between(q['round'], q['min'], q['max'], color=line.get_color(), alpha=0.1, label='Range')


# plt.plot(q1['round'], q1['mean_score'])
# plt.fill_between(q1['round'], q1['min'], q1['max'], color='blue', alpha=0.1, label='Range')

# plt.plot(q2['round'], q2['mean_score'])
# plt.fill_between(q2['round'], q2['min'], q2['max'], color='orange', alpha=0.1, label='Range')

# plt.plot(q3['round'], q3['mean_score'])
# plt.fill_between(q3['round'], q3['min'], q3['max'], color='green', alpha=0.1, label='Range')


plt.ylim(-0.5, 3.0)

plt.xlabel('Edit rounds')
plt.xticks(range(0, 16))

plt.show()
```

```python
fig, axs = plt.subplots(4, 9, figsize=[30, 10])

for i, question in enumerate(ordered['question'].unique()):
    q = ordered[ordered['question'] == question][['round', 'mean_score', 'max', 'min']]

    line, = axs[i // 9, i % 9].plot(q['round'], q['mean_score'], alpha=1.0, linewidth=1, marker='o', markersize=3)
    # plt.fill_between(q['round'], q['min'], q['max'], color=line.get_color(), alpha=0.1, label='Range')
    # axs[i // 4, i % 4].set_title(question)

    axs[i // 9, i % 9].set_xlim((0, 16))
    axs[i // 9, i % 9].set_ylim((-0.5, 3.0))


plt.subplots_adjust(wspace=0.2)
plt.show()
```
